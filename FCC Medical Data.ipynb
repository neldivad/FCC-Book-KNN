```python
import pandas as pd
import numpy as np
```

```python
PATH = "C://Users/User/Desktop/"
df = pd.read_csv(PATH + "medical_examination.csv")
df
```

```python
# Add BMI
df['BMI'] = df['weight'] / (df['height'] / 100) **2
df
```

```python
# Add overweight
df['overweight'] = df['BMI'].apply(lambda x: 1 if x > 25 else 0)
df
```

```python
df['cholesterol'] = df['cholesterol'].apply(lambda x: 0 if x <= 1 else 1)
df
```

```python
df['gluc'] = df['gluc'].apply(lambda x: 0 if x <= 1 else 1)
df
```

```python
import seaborn as sns
value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight']
long = pd.melt(df, id_vars=['cardio'], value_vars= value_vars,)

long
```

```python
long = pd.DataFrame( long.groupby(['variable', 'value', 'cardio'])['value'].count() ) \
    .rename( columns={'value': 'total'} ) \
    .reset_index() # wow magic line

long
```

```python
sns.catplot(data= long, x= 'variable', y='total', hue= 'value',
            col= 'cardio', kind= 'bar')
```

## Filter Incorrect Data

```python
# Filter incorrect data
df[ df['ap_lo'] <= df['ap_hi'] ]
```

```python
# bottom 0.025% in height
df[ df['height'] >= df['height'].quantile(0.025) ]
```

```python
# top 0.975 in height
df[ df['height'] <= df['height'].quantile(0.975) ]
```

```python
df[ df['weight'] >= df['weight'].quantile(0.025) ]
```

```python
df[ df['weight'] <= df['weight'].quantile(0.975) ]
```

```python
df = df[ df['ap_lo'] <= df['ap_hi'] ]
df = df[ df['height'] >= df['height'].quantile(0.025) ]
df = df[ df['height'] <= df['height'].quantile(0.975) ]
df = df[ df['weight'] >= df['weight'].quantile(0.025) ]
df = df[ df['weight'] <= df['weight'].quantile(0.975) ]
df
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 6))
correlation = df.corr()

# creating mask
mask = np.triu(np.ones_like(df.corr()))

sns.heatmap(correlation, mask = mask, cmap = 'Greys')
```
