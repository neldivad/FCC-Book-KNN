```python
def add_time(start_time, time_to_add, starting_day= None):
    ##################
    # Fail test
    ##################
    regexed = ""
    starting_days = ['monday',
                     'tuesday',
                     'wednesday',
                     'thursday',
                     'friday',
                     'saturday',
                     'sunday',
                     None]
    
    if starting_day == None:
        pass
    else:
        regexed = starting_day.lower()
        if regexed not in starting_days:
            raise ValueError("Invalid starting day. Expected one of: %s" % starting_days)

    accepts = '0123456789: APM'
    
    for char in start_time and time_to_add:
        if char not in accepts: 
            return "Error: Accepted input is only limited to numbers, meridium, and ':'."
    
    start = start_time.split(' ')[0]
    meridium = start_time.split(' ')[1]
    if meridium == 'AM' or meridium == 'PM':
        pass
    else:
        return "Error: Accepted meridiums are 'AM' or 'PM'."
    
    start_hours = int( start.split(':')[0] )
    start_minutes = int( start.split(':')[1] )
    hours_to_add = int( time_to_add.split(':')[0] )
    minutes_to_add = int( time_to_add.split(':')[1] )
    
    if start_hours <1 or start_hours > 12:
        return "Error: Starting hours incorrectly specified"
    if start_minutes <0 or start_minutes > 59:
        return "Error: Starting minutes incorrectly specified"
    if minutes_to_add > 59:
        return "Error: Minutes to add incorrectly specified"
    
    ######################
    # Calculate 
    ######################
    if meridium == 'PM' and start_hours != 12:
        start_hours += 12

    all_minutes = start_minutes + minutes_to_add
    minutes_left = all_minutes % 60
    
    extra_hours = all_minutes // 60
    all_hours = start_hours + hours_to_add + extra_hours
    ending_hour = all_hours % 24
    days_passed = all_hours //  24
    
    if ending_hour > 12:
        end_meridium = 'PM'
        ending_hour -= 12
    elif ending_hour == 12:
        end_meridium = 'PM'
    else:
        end_meridium = 'AM'
    
    # Edge case
    if ending_hour == 0:
        ending_hour += 12
    
    # Single digit minute fix
    if minutes_left < 10:
        minutes_left = '0' + str(minutes_left)
        
    #########################
    # Print if no day input
    #######################
    if starting_day == None:
        if days_passed == 0:
            last_string = ""
        elif days_passed == 1:
            last_string = " (next day)"
        else:
            last_string = " ({} days later)".format(days_passed)
        return ('{}:{} '.format(ending_hour, 
                                       minutes_left) 
              + end_meridium
              + last_string
             )
    
    else:
        weekday_counter = 0
        if regexed == 'monday':
            weekday_counter = 0
        elif regexed == 'tuesday':
            weekday_counter = 1
        elif regexed == 'wednesday':
            weekday_counter = 2
        elif regexed == 'thursday':
            weekday_counter = 3
        elif regexed == 'friday':
            weekday_counter = 4 
        elif regexed == 'saturday':
            weekday_counter = 5
        elif regexed == 'sunday':
            weekday_counter = 6

        day_counter = (days_passed + weekday_counter) % 7 

        if day_counter == 0:
            ending_day = 'Monday'
        elif day_counter == 1:
            ending_day = 'Tuesday'
        elif day_counter == 2:
            ending_day = 'Wednesday'
        elif day_counter == 3:
            ending_day = 'Thursday'
        elif day_counter == 4:
            ending_day = 'Friday'
        elif day_counter == 5:
            ending_day = 'Saturday'
        elif day_counter == 6:
            ending_day = 'Sunday'
            
        if days_passed == 0:
            last_string = ", " + ending_day
        elif days_passed == 1:
            last_string = ", " + ending_day + " (next day)"
        else:
            last_string = ", " + ending_day + " ({} days later)".format(days_passed)
        return('{}:{} '.format(ending_hour, 
                                       minutes_left) 
              + end_meridium
              + last_string
             )
```

```python
print( add_time("3:30 PM", "2:12") ) 
print( add_time("11:55 AM", "3:12") ) 
print( add_time("9:15 PM", "5:30") )
print( add_time("11:40 AM", "0:25") )
print( add_time("2:59 AM", "24:00") )
print( add_time("11:59 PM", "24:05") )
print( add_time("8:16 PM", "466:02") )
print( add_time("5:01 AM", "0:00") )
print( add_time("3:30 PM", "2:12", "Monday") )
print( add_time("2:59 AM", "24:00", "saturDay") )
print( add_time("11:59 PM", "24:05", "Wednesday") )
print( add_time("8:16 PM", "466:02", "tuesday") )
```
