```python
import pandas as pd
import numpy as np
```

```python
PATH = "C://Users/User/Desktop/"
df = pd.read_csv(PATH + "adult.data.csv")
df
```

```python
# How many of each race are represented in this dataset?
df["race"].value_counts()
```

```python
# What is the average age of men?
round( df[ df['sex'] == "Male" ]['age'].mean() , 1) 
```

```python
# What is the percentage of people who have a Bachelor's degree?
round( df['education'].value_counts(normalize=True)[2] * 100, 1)
```

```python
# with and without 'Bachelors', 'Masters', or 'Doctorate'
adv = df.loc[ df.education.isin(['Bachelors','Masters', 'Doctorate']) ] 
nadv = df.loc[~df.education.isin(['Bachelors','Masters', 'Doctorate']) ]
```

```python
# percentage with salary >50K
round( adv['salary'].value_counts(normalize=True)[1] *100, 1)
```

```python
# percentage with salary >50K
round( nadv['salary'].value_counts(normalize=True)[1] *100, 1)
```

```python
# What is the minimum number of hours a person works per week (hours-per-week feature)?
min_work_hours = df['hours-per-week'].min()
min_work_hours
```

```python
# What is the minimum number of hours a person works per week
# min_hour = df.loc[ df['hours-per-week'].isin(['1']) ]
num_min_workers = len(df[df['hours-per-week']== min_work_hours])
num_min_workers
```

```python
# What percentage of the people who work the minimum number of hours per week have a salary of >50K?
rich_perc = round(len(df[(df['hours-per-week'] == min_work_hours) \
                         & (df['salary'] == '>50K')]) / num_min_workers * 100, 1)
rich_perc
```

```python
# What country has the highest percentage of people that earn >50K?
country_income = df[ ['salary', 'native-country'] ]
all_country_income = country_income.groupby('native-country').count()

all_country_income
```

```python
plus_50k = country_income[ country_income['salary'] == '>50K' ]
plus_50k = plus_50k.groupby('native-country').count()
plus_50k
```

```python
plus_50_perc = (plus_50k / all_country_income) *100
plus_50_perc.sort_values(by=['salary'], inplace=True, ascending= False)

print(plus_50_perc.index[0])
print( round(plus_50_perc['salary'][0], 1) )
```

```python
# Identify the most popular occupation for those who earn >50K in India.
rich_indian = df[ (df['salary'] == '>50K') & (df['native-country'] == 'India') ]
rich_indian['occupation'].value_counts().index[0]
```
